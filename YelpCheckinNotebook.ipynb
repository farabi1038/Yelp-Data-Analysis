```python
%%R -w 1000 -h 1000

# try to see if we can predict stars from checkin times
data_bus <- read.csv("yelp_academic_dataset_business.csv", stringsAsFactors = TRUE)
data_checkin<- read.csv("yelp_academic_dataset_checkin.csv",stringsAsFactors=TRUE)
data_subset = subset(data_bus,select=c("stars","business_id"))
data_checkin_stars <- merge(data_checkin,data_subset,by="business_id")

cat("Dimension:\n"); print(dim(data_checkin_stars))
print(names(data_checkin_stars))
cat("Dimension:\n"); print(dim(data_bus))
print(names(data_bus))
# used so we can use weka to analyze
write.csv(data_checkin_stars, file = "yelp_checkins_stars.csv") 
```

```python
%%R
print(dim(data_checkin_stars))
#print(head(data_checkin_stars))
##PREPROCESSING
drops <- c("type","business_id") # dropping columns with only 1 factor
data_checkin_stars = data_checkin_stars[,!(names(data_checkin_stars) %in% drops)]
data_checkin_stars[is.na(data_checkin_stars)] <- 0 #replace all NA values with 0
data_checkin_stars$stars = round(data_checkin_stars$stars) #round to nearest int



##SUBSAMPLING 1000 out of 100000 (training set)
subsample_indices <- sample(nrow(data_checkin_stars), 9000)
subsample <- data_checkin_stars[subsample_indices, ]

#SUBSAMPLING FOR TESTING SET

testset_indices <- data_checkin_stars[~subsample_indices, ]


##ROUGH GLM MODEL TO SEE IF CHECK INS HAVE ANY RELATIONSHIP WITH STARS
checkin.glm <- glm(stars~.,data=subsample,na.action=na.omit)
predictions <- predict(checkin.glm , subsample)
predictions <- round(predictions)
incorrect.prediction <- (predictions != subsample$stars)
hist(subsample$stars, xlab="Subsample Star Dist", ylab="Frequency", main="Sample Frequency", breaks=6, col="blue", prob=TRUE)
hist(predictions, xlab="Prediction Star Dist", ylab="Frequency", main="Sample Frequency", breaks=6, col="blue", prob=TRUE)
plot(checkin.glm)
m <- nrow(subsample_1000)
print(m)

cat("Summary of sample set:\n"); print(summary(subsample_1000$stars))
cat("Summary of predictions (on training set):\n"); print(summary(predictions))

# Confusion matrix
# Very heavily biased towards 4 star ratings, due to class imbalance, 
# next step is to build a model with more even prior class distributions 
# and actually try it on a testing set
confusion_matrix <- table(subsample$stars, predictions)
cat("Confusion matrix: "); print(confusion_matrix); cat("\n")

accuracy <- (m - sum(incorrect.prediction))/m ## ~ 63% accuracy on training set (beats random guessing by 40%)
cat("Accuracy: "); print(accuracy) 
```
