```python
%%R
data_bus <- read.csv("yelp_academic_dataset_business.csv", stringsAsFactors=FALSE)
data_checkin<- read.csv("yelp_academic_dataset_checkin.csv",stringsAsFactors=FALSE)

```

```python
%%R

# Fit a distribution to star values among restaurants
loc <- grep("R,e,s,t,a,u", data_bus$categories, perl=TRUE, value=FALSE)
data_bus_subset <- data_bus[loc,]
hist(data_bus$stars, xlab="Star Dist among Restaurants", ylab="Frequency"
     , main="Sample Frequency", breaks=12, col="red", prob=TRUE)


data_subset = subset(data_bus_subset,select=c("stars","business_id"))
data_checkin_stars <- merge(data_checkin,data_subset,by="business_id")

library(MASS)
star_dist <- fitdistr(data_bus$stars, "normal")
print(star_dist$estimate)
print(summary(data_bus$stars))
curve(dnorm(x, mean=star_dist$estimate[1], sd=star_dist$estimate[2]), col="blue", add=TRUE)
legend("right", c("Normal Dist Fit"), col=c("blue"), lwd=3)

cat("Dimension of rest checkin with stars:\n"); print(dim(data_checkin_stars))
#print(names(data_checkin_stars))
# used so we can use weka to analyze
data_checkin_stars[is.na(data_checkin_stars)] <- 0 #replace all NA values with 0
write.csv(data_checkin_stars, file = "yelp_checkins_stars.csv") 
```

##Star Distribution Among Restaurants
We found that a normal distribution fitted the distribution of stars among restaurants very nicely, with a mean of 3.67 and a s.d. of .9.  Furthermore, the median was 3.5 which suggests that the distribution is not too skewed and the normal is a good fit.  This is interesting because one would expect a normal distribution centered around 2.5, to define an "average" restaurant, when in reality, an "average" restaurant actually has a rating of 3.5.  A standard deviation of 1 signifies by the 68/95 rule that a resturant rated 4.5 or higher falls in the top 5% of total resturants, while a restaurant rated 1.5 or lower falls in the bottom 5 percentile of resturants.  A normal distribution suggests that each restaurant's star is independently and identically distributed, and gives us the insight that one restaurant's rating has little impact on another's rating, and things such as location should not affect a restaurants yelp rating.

```python
%%R
# Try to see if we can seperate "good" from "bad" restaurants
# by using check-in info.  We define "good" as receiving a rating
# above the median, and "bad" as having a rating below the median.

print(median(train_checkin$stars))
good_rest<- train_checkin$stars > median(train_checkin$stars)

train_rating<-train_checkin
train_rating$rating<-0
train_rating[good_rest,]$rating<-1 #assign 0 or 1 value depending on "good" or "bad"

train_rating$stars <-NULL #drop the stars column because we do not want to predict on it

train_rating$rating <- factor(train_rating$rating) # turn good and bad into factors
print(summary(train_rating$rating))
print(names(train_rating))
```

```python
%%R

good.lda <- lda(rating~.,data=train_rating,na.action=na.omit)
plot(good.lda)
print(good.lda)
training_predictions1 <- predict(good.lda , train_rating)
m <- nrow(train_rating)
print(m)
cat("Summary of sample set:\n"); print(summary(train_rating$rating))
cat("Summary of predictions (on training set):\n"); print(summary(training_predictions1$class))
incorrect.prediction <- (training_predictions1$class != train_rating$rating)
confusion_matrix <- table(train_rating$rating, training_predictions1$class)
cat("Confusion matrix: "); print(confusion_matrix); cat("\n")

accuracy <- (m - sum(incorrect.prediction))/m
cat("Accuracy: "); print(accuracy)
```
