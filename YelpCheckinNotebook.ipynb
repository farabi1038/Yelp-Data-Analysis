## Investigation into Yelp Restaurant Star Ratings / Check-in Times
The goal of this notebook was primarily to:
- fit distributions to check in times / overall star ratings, along with a description of what natural processes generated these distributions, offering insight into the way people check in / rate restaurants. 
- investigate if there was any correlation between Restaurant check-in times and star ratings for restaurants.
- build a predictive model of restaurant stars/sentiment given check-in times.  
    
Using this information, restaurants could possibly offer incentives for visiting them during these influential check-in times, or focus on improving service during these hours, improving their overall yelp rating.
    
Users of yelp would benefit from analyses of distributions of check in times / star ratings.  A star rating by itself is has little meaning without some sort of distribution backing it up.  For example if every restaurant was rated 5 on yelp, then a 5 rated restaurant would not be a relatively good restaurant. Relative "goodness" is a more useful metric to a yelp user, which we can characterize after we attempt to fit a distribution to yelp restaurant stars.

```python
%%R
data_bus <- read.csv("yelp_academic_dataset_business.csv", stringsAsFactors=FALSE)
data_checkin<- read.csv("yelp_academic_dataset_checkin.csv",stringsAsFactors=FALSE)
```

##Preprocessing Steps
The dataset we used for checkin primarily came from the provided yelp_academic_dataset_checkin.json file, however we had to do several preprocessing steps.  
1. Filter out all non-restaurant related check-in times.
2. Replace N/A check-in values with 0.  We are assuming that yelp did not fail to collect any information, so our conclusion was that a N/A value means that there were no check ins for that particular time period for that particular restaurant.
3. Merge with business dataset based on business_id to add **stars** feature column to dataset.
4. Aggregate data across check-in times and add that as a feature column, **total_checkins** per restaurant.
5. Add an additional column factor **quant_rating**, which was "high"(top 25%), "med"(middle 50%) or "low"(bottom 25%)  depending on which quartile a restaurant fell into in the overall distribution.  This feature was created because star prediction yielded very low accuracy.  We needed an easier problem to solve rather predicting stars that had even class sizes e.g. "high" vs "low", and quartiles fit that description well.  **Note: we were careful to not predict stars out of features we constructed on stars, e.g. "quant_rating"**

**Final Format of Data**

After cleaning up our data, we had a 5556 x 171 matrix, where:

1. **Rows** represented each restaurant.
2.  One additional row was added, which was aggregation of total_checkins_time per time slice across Restaurants.  
3. **Columns** represented a check_in time for a particular hour/day of the week, in the form of **checkin_info_18.5** where 18 indicates the hour (6 pm) and 5 represents the day of the week (Saturday), from this we have 7*24 or 168 columns.  
4. Three Additional Columns were added, **total_checkins_rest** for a restaurant along with its **star** rating(1-5), and overall** quant_rating**("high", "med","low"). 

```python
%%R
loc <- grep("R,e,s,t,a,u", data_bus$categories, perl=TRUE, value=FALSE)
data_bus_subset <- data_bus[loc,]
data_subset = subset(data_bus_subset,select=c("stars","business_id"))
data_checkin_stars <- merge(data_checkin,data_subset,by="business_id")
drops <- c("type","business_id")
data_checkin_stars = data_checkin_stars[,!(names(data_checkin_stars) %in% drops)]
data_checkin_stars[is.na(data_checkin_stars)] <- 0 #replace all NA values with 0

first_q = 3.01 # assign the label "good" to nodes in top quantile of our data, and similar "bad" for the lowest quantile
third_q = 3.99
good<- data_checkin_stars$stars > third_q
bad<- data_checkin_stars$stars < first_q


data_checkin_stars$quant_rating <- "medium"
data_checkin_stars[good,]$quant_rating<-"good"
data_checkin_stars[bad,]$quant_rating<-"bad"
print(dim(data_checkin_stars))
```

```python
%%R -h 400 -w 800
## Fitting a distribution to overall star rating per restaurant
## Originally we settled on normal as being a good fit.  Seeing as the tails are fatter than expected, we also tried to fit 
# a Cauchy dist. to our data.
library(MASS)
data_checkin_stars_only <- data_checkin_stars[1:169]
star_dist <- fitdistr(data_checkin_stars_only$stars, "normal")
star_dist1 <- fitdistr(data_checkin_stars_only$stars, "cauchy")
par(mfrow = c(1, 2))
hist(data_checkin_stars_only$stars, xlab="Stars", ylab="Frequency",
     breaks=13, main="Star Distribution among Restaurants", col="red", prob=TRUE)
curve(dcauchy(x, location=star_dist1$estimate[1] , scale=star_dist1$estimate[2]), col="green", add=TRUE)
curve(dnorm(x, mean=star_dist$estimate[1], sd=star_dist$estimate[2]), col="blue", add=TRUE)
legend("right", c("Normal Dist Fit","Cauchy Fit"), col=c("blue","green"), lwd=3)
cat("Quantile summary for star rating among Restaurants: \n");print(summary(data_bus_subset$stars))
ylim <- c(0,max(data_checkin_stars_only$stars))
Cauchy_dist_quartiles <- rcauchy(10000, location=star_dist1$estimate[1] , scale=star_dist1$estimate[2])
normal_dist_quartiles <- rnorm(10000, mean=star_dist$estimate[1], sd=star_dist$estimate[2])

qqplot(data_checkin_stars$stars, Cauchy_dist_quartiles,xlab="Cauchy-Fit QQ Plot"
       ,ylim=ylim,ylab="Sampled Quartiles",main = "Cauchy QQ Plot")
abline(0,1)
qqplot(data_checkin_stars$stars, normal_dist_quartiles,xlab="Normal-Fit QQ Plot"
       ,ylim=ylim,ylab="Sampled Quartiles",main = "Normal QQ Plot")
abline(0,1)
```

##Star Distribution Among Restaurants
We characterized the star distribution as **normal, with MLE estimates of the mean as 3.46 with a s.e. of 0.73**.  The quartiles matching up on the QQ plot, small standard errors for parameter estimates, and the fairly symmetrical distribution(median = 3.5,  mean=3.46) all signify that N~(3.46,.73) is a good fit.  Attempts to fit a distribution with a fatter tail, e.g. Cauchy, yielded a similar, but worse fit overall along with higher avg. errors, and worse QQ-plot fits.  Furthermore, the Cauchy distribution would sometimes predict numbers such as -32 which could never possibly be in the range of the 1-5 star rating yelp system.  

This is interesting because we expected the average restaurant to get a review of 3 (the average of 1 and 5), but it is in actually 3.5, suggesting people tend to rate restaurants higher than they are "supposed to be rated".  Furthermore, by using properties of the normal distribution, namely the 3 sigma rule, we can say that stars roughly translate into these percentiles:

##Fitting distributions to check in times

```python
%%R -h 700 -w 700
par(mfrow=c(2, 2))
raw_checkin_times <- data_checkin_stars[,1:168]

total_checkin_times <- colSums(raw_checkin_times)
total_checkin_sorted_names = names(data_checkin_stars)
ndx = order(total_checkin_times,decreasing=TRUE)

total_checkin_sorted_names<-total_checkin_sorted_names[ndx]
total_checkin_times <- total_checkin_times[ndx]

plot(total_checkin_times,log='xy',main="Ranked Log-Log Plot")
plot(total_checkin_times,log='x', main= "Ranked Log-Lin Plot") # Log-Log and Log-lin plot checking for log-normal dist
plot(total_checkin_times,log='y', main= "Ranked Lin-Log Plot") # Log-Log and Log-lin plot checking for log-normal dist
gamma_param = fitdistr(total_checkin_times/10, "Gamma") # attempt to fit gamma, exp, and log norm
exp <-fitdistr(total_checkin_times,"exponential")
lognorm_param <-fitdistr(total_checkin_times,"lognormal")

hist(total_checkin_times, xlab="Check ins", ylab="Frequency",
     breaks=20, main="Total number of check ins in a given time slice", col="green", prob=TRUE)
legend("right", c("Exponential Distribution","Gamma Distribution", "Log-Normal"), col=c("blue", "red", "black"), lwd=3)

x_range = 1:length(total_checkin_times)
curve(dexp(x, rate=exp$estimate[1]), col="blue", add=TRUE)
curve(dlnorm(x, meanlog=lognorm_param$estimate[1], sdlog=lognorm_param$estimate[2]), col="black", add=TRUE)
curve(dgamma(x, shape=gamma_param$estimate[1], rate=gamma_param$estimate[2]/10), col="red", add=TRUE)
cat("KS p-value exp: ");print((ks.test(total_checkin_times,"pexp",rate=exp$estimate[1]))$p.value) #compare KS stats
cat("KS p-value lognorm: ");print((ks.test(total_checkin_times,"plnorm",meanlog=lognorm_param$estimate[1], sdlog=lognorm_param$estimate[2]))$p.value)
cat("KS p-value gamma: ");print((ks.test(total_checkin_times,"pgamma",shape=gamma_param$estimate[1], rate=gamma_param$estimate[2]/10))$p.value)


n <- length(total_checkin_times)
x <- qexp(c(1:n)/(n+1))
m <- mean(total_checkin_times)
ylim <- c(0,max(total_checkin_times))
qqplot(x, total_checkin_times, xlab="Exp QQ Plot",ylim=ylim,ylab="Sampled Quartiles",main="Exp QQ Plot")
abline(0,m,lty=2)


qqnorm(log(total_checkin_times),main="Log-normal QQ plot")
qqline(log(total_checkin_times)) 

gamma_dist_quartiles <- rgamma(1000, shape=gamma_param$estimate[1], rate=gamma_param$estimate[2]/10)
qqplot(total_checkin_times, gamma_dist_quartiles,xlab="Gamma QQ Plot",ylim=ylim,ylab="Sampled Quartiles",main = "Gamma QQ Plot")
abline(0,1)
```

##Check in Distribution 
Judging from the histogram of total checkins per time slice (one hour), the distribution seems to follow either a **log-normal, exponential, or gamma distribution**.  We can rule out log-normal by plotting the log-log plot of the ranked data, and seeing that it is clearly not linear, and further verifying this by showing that both the log-lin and the lin-log plots are quite linear, suggesting an exponential relationship.  The exponential distribution describes the time in between events and through a simple transformation can describe the number of times an event occurs in a particular time slice.  Since we are taking the total checkins, or **summing up independently distributed exponentially distributed variables, by definition this describes the Gamma Distribution.**  Furthermore, from the QQ plots we can see that the gamma QQ plot is almost a perfect fit until we reach roughly the third quartile.  The sharp change in slope indicates that there reaches a point where the distribution changes from a gamma to one with fatter tails, perhaps reflecting a power law for the top ~ 25 percentile of check in times.  Our hypothesis was further validated by performing a KS test in which the gamma distr with MLE fit parameters had a much higher p-value than its log-normal and exponential counterparts.  Next we investigate which days are the most popular, by printing the first couple of elements in our ranked list and also plotting a ranking of restaurants.

```python
%%R -h 700 -w 700

##Building a predictive model to predict star rating based on checkin_info using logistic regression.
##SUBSAMPLING 3000 out of 4000 (training set)
train_checkin_indices <- sample(nrow(data_checkin_stars), 3000)
train_checkin <- data_checkin_stars[train_checkin_indices, ]
write.csv(train_checkin, file = "train_yelp_checkins_stars.csv") 

#SUBSAMPLING FOR TESTING SET (all elements not in training set)
test_checkin <- data_checkin_stars[-train_checkin_indices, ]
write.csv(test_checkin, file = "test_yelp_checkins_stars.csv")

##ROUGH GLM MODEL TO SEE IF CHECK INS HAVE ANY RELATIONSHIP WITH STARS
no_rating <- train_checkin[,1:169]
checkin.glm <- glm(stars~.,data=no_rating,na.action=na.omit)

test_predictions <- predict(checkin.glm , test_checkin)
training_predictions <- predict(checkin.glm , train_checkin)

total_error <- abs(test_predictions - test_checkin$stars)
cat("Quartile summary of abs error");print(summary(total_error))

par(mfrow = c(2, 2))
hist(test_checkin$stars, xlab="Test Sample Star Dist", ylab="Frequency"
     , main="Test Sample Frequency", breaks=6, col="blue", prob=TRUE)
hist(test_predictions, xlab="Test Prediction Star Dist", ylab="Frequency"
     , main="Test Set Prediction Frequency", breaks=6, col="blue", prob=TRUE)

hist(train_checkin$stars, xlab="Training Set Prediction Star Dist", ylab="Frequency"
     , main="Training Set Prediction Frequency", breaks=6, col="red", prob=TRUE)
hist(training_predictions, xlab="Training Set Prediction Star Dist", ylab="Frequency"
     , main="Training Set Prediction Frequency", breaks=6, col="red", prob=TRUE)
plot(checkin.glm)
m <- nrow(test_checkin)
cat("On average we have an absolute error in predicting star ratings of: "); print(sum(total_error)/m);
```

## Use Of Weka
In the above cell I exported the training and testing data in csv form so I could use Weka and run a bunch of classifiers on the star data, and automatically perform 10 cross fold CV without typing much code.  Because this dataset was much smaller than the reviews and user dataset, it was possible to import the entire dataset and run a variety of classifiers.   After running Random Forest, SVM, Logistic Regression, Adaboost, LDA, QDA,  none of them saw abs. errors significantly lower than the from the GLM model.  Furthermore, attempts to use Weka's feature selection tool led to a different set of features every time; meaning even if a star classifier could be built, it would not be robust and lead to different results depending on how lucky we got.  Based on these issues, I believe that there is not enough information / check-in info to make accurate predictions of star category.  To simplify the problem, I reduced the original star classification problem to a binary classification problem where we try to predict whether a restaurant is "good" or "bad", meaning that it falls into the bottom or top quartile based on check in info.

```python
%%R
#print(names(train_checkin))

data_checkin_stars$stars <- NULL
drop_indices <- data_checkin_stars$quant_rating=="medium"
binary_dataset <- data_checkin_stars[!drop_indices,] # now a 3600 dim dataset
write.csv(binary_dataset, file = "binary_dataset.csv")

train_checkin_indices <- sample(nrow(binary_dataset), 2300)
train_checkin <- binary_dataset[train_checkin_indices, ]
test_checkin <- binary_dataset[-train_checkin_indices, ]

qda_model <- qda(quant_rating ~ ., data=train_checkin)
predictions <- predict(qda_model, test_checkin)
#print((predictions$class))
incorrect.prediction <- (predictions$class != test_checkin$quant_rating)

m <- nrow(test_checkin)
confusion_matrix <- table(test_checkin$quant_rating, predictions$class)
cat("Confusion matrix: "); print(confusion_matrix); cat("\n")

accuracy <- (m - sum(incorrect.prediction))/m
cat("Accuracy: "); print(accuracy)
```

```python
%%R

lda_model <- lda(quant_rating ~ ., data=train_checkin)
predictions <- predict(lda_model, test_checkin)
incorrect.prediction <- (predictions$class != test_checkin$quant_rating)

m <- nrow(test_checkin)
confusion_matrix <- table(test_checkin$quant_rating, predictions$class)
cat("Confusion matrix: "); print(confusion_matrix); cat("\n")

accuracy <- (m - sum(incorrect.prediction))/m
cat("Accuracy: "); print(accuracy)
```

##Conclusions and Results
1. Restaurant star reviews are distributed N~(3.46,.73), but with slightly heavier tails than the normal distribution would explain.  This is not too big of a problem since we are restricted in our star rating domain in between 1-5 stars and in our distribution over 99% of the points fall within these boundaries.

2. Aggregation of check ins by hour results in a Gamma RV which represents the distribution of the number of check ins in any particular hour, except for a special subset of hours, namely 4-5 on Friday, Saturday, and Sunday, which seem to follow a power law or exhibit more extreme values than the gamma distribution would predict.  Most importantly, we garnered from the data that Friday, especially the hours of noon, 5, 6, and 7, are responsible for 10% of a Restaurants check-ins and potentially clientele.  

3. Using a crude logistic regression model we were able to predict star reviews within a factor of 0.5411736 on average using check-in info.  After examining the testing distribution and predicted values though,  our classifier merely classifies every instance as a "3.5" restaurant, minimizing test error but not giving us any predictive power.  Furthermore, since the s.d. is .75 we could be off by a percentile of 20-30% depending on how close we are to the mean, on predicting the quality of a restaurant.   

4. Using LDA and labeling the top quantile of check in hours as "good" and the bottom quantile as "bad", we were able to predict classes with 0.64 accuracy using a QDA classifier vs .60 using a LDA classifier.  Neither of these is that much of an improvement on random guessing, and the PCA plot shows that it is hard to find seperating boundaries between the classes; meaning either we do not have enough data or we have not established a proper model for predicting restaurant quality given check in info.  One property of check in info which is interesting was the high temporal variance or sinusoidal pattern we observed in the time series graph.  Due to the repeating pattern of the underlying data, it is difficult to create a linear classifier for data which does not also follow this pattern, if not impossible.  

5. Given the limited amount of time we had to work on three datasets, we did not manage to find a good classification method relating star rating and check in info.  However, we did manage to uncover some patterns among yelp users including categorizing the check-in time distribution, seeing what quartile a star corresponded to, and uncovering the most popular check in times.
